## Regresi Linear antara masing-masing variabel pembentuk beton & kuat tekan beton


### Regresi Linear antara semen (cement) & kuat tekan beton (csMPa)

```
b1_cement = (len(data)*sum(cement*csmpa)-(sum(cement)*sum(csmpa)))/ (len(data)*sum(cement**2)-(sum(cement)**2))
b0_cement = csmpa.mean() - b1_cement*cement.mean()

print('Î²0 =', format(b0_cement, '.4f'))
print('Î²1 =', format(b1_cement, '.4f'))
```

```
y_cement = b0_cement + b1_cement*cement
data.plot(kind='scatter', x='cement', y='csMPa', alpha=0.5, c='indianred')
plt.plot(cement, y_cement, "k", linewidth=3)
plt.title('Regresi Linear Semen dengan Kuat Tekan Beton', fontsize=14)
```

```
SSE_cement = sum((csmpa - y_cement)**2)
print('Sum Squared Error (SSE) =', format(SSE_cement, '.4f'))
```

```
import math

variansi_error_cement = SSE_cement/(len(cement)-2)
s_cement = math.sqrt(variansi_error_cement)
print('Variansi Error =', format(variansi_error_cement, '.4f'))
```

```
koef_determinasi_cement = 1 - (SSE_cement/sum((csmpa-cement.mean())**2))
print('Koefisien Determinasi =', format(koef_determinasi_cement, '.4f'))
```

```
Sxx_cement = sum((cement-cement.mean())**2)

# Interval kepercayaan slope (Î²1)
t_0025 = 2.045
batas_bawah_slope_cement = b1_cement - t_0025*(s_cement/math.sqrt(Sxx_cement))
batas_atas_slope_cement = b1_cement + t_0025*(s_cement/math.sqrt(Sxx_cement))
print(format(batas_bawah_slope_cement, '.4f'),"< Î²1 <",format(batas_atas_slope_cement, '.4f'))

print("")

# Interval kepercayaan intercept (Î²0)
t_0025 = 2.045
batas_bawah_intercept_cement = b0_cement - t_0025*(s_cement/math.sqrt(len(cement)*Sxx_cement))*math.sqrt(sum(cement**2))
batas_atas_intercept_cement = b0_cement + t_0025*(s_cement/math.sqrt(len(cement)*Sxx_cement))*math.sqrt(sum(cement**2))
print(format(batas_bawah_intercept_cement, '.4f'),"< Î²0 <",format(batas_atas_intercept_cement, '.4f'))
```

```
t_0025 = 2.045
batas_bawah_cement = []
batas_atas_cement = []
mean_x_y_cement = []

for i in range(0,len(data)):
    b_b_cement = (y_cement[i]) - t_0025*s_cement*math.sqrt((1/len(data))+(cement[i]-cement.mean())**2/Sxx_cement)
    batas_bawah_cement.append(b_b_cement)
    b_a_cement = (y_cement[i]) + t_0025*s_cement*math.sqrt((1/len(data))+(cement[i]-cement.mean())**2/Sxx_cement)
    batas_atas_cement.append(b_a_cement)
    mean_x_y_cement.append((b_a_cement-b_b_cement)/2 + b_b_cement)
```

```
from matplotlib.pyplot import figure
figure(figsize=(7, 5), dpi=100)

plt.scatter(cement,csmpa,s=7,c='indianred',label='csMPa')
plt.plot(cement,batas_atas_cement,c='blue',label='Batas Atas & Batas Bawah Rataan Y|x, CI = 95%')
plt.plot(cement,batas_bawah_cement,c='blue')
plt.plot(cement,mean_x_y_cement,c='red',label='Rataan Batas Atas & Batas Bawah Y|x, CI = 95%')

plt.title('Interval Kepercayaan Rataan csMPa diketahui cement')
plt.xlabel('cement')
plt.ylabel('csMPa')

plt.legend()
```

**Hipotesis**<br>
Hipotesis Null ($H_0$): saat $x_0$ = 200kg, maka $Î¼$ kuat tekan beton â‰¤ 60 MPa<br>
Hipotesis Alternatif ($H_1$): saat $x_0$ = 200kg, maka $Î¼$ kuat tekan beton > 60 MPa

**Menentukan Signifikansi**

```
import scipy.stats as z

significance = 0.05
critical_val = z.norm.ppf(q = 1 - significance)
print('Critical Value =', format(critical_val, '.4f'))
```

```
from numpy import arange
from scipy.stats import norm

x_axis = arange(-3, 3, 0.00001)
y_axis = norm.pdf(x_axis, 0, 1)

plt.vlines([0], 0, 0.4, linestyles='dashed', colors='black', label="Z Populasi = 0")
plt.vlines([1.645], 0, 0.1, colors='purple', label="Z Critical Value = 1.645")
plt.annotate('1.645', xy=(1.5,0.14), fontsize=15)
plt.title("Distribusi Kuat Tekan Beton", fontsize=16)
plt.plot(x_axis, y_axis)
plt.legend()
plt.show()
```

**Menghitung P-Value dari kuat tekan beton**

```
y_cement_200 = b0_cement + b1_cement*200
z_score_csmpa = (y_cement_200-60) / (csmpa.std()/(len(csmpa))**0.5)
print('Z score kuat tekan beton = ', format(z_score_csmpa,'.4f'))
```

```
x_axis = arange(-3, 3, 0.00001)
y_axis = norm.pdf(x_axis, 0, 1)

plt.style.use('seaborn')
plt.vlines([0], 0, 0.4, linestyles='dashed', colors='black', label="Z Populasi = 0")
plt.vlines([1.645], 0, 0.1, colors='purple', label="Z Critical Value = 1.645")
plt.vlines([z_score_csmpa], 0, 0, colors='cadetblue', label="Z kuat tekan beton = -58.8656")
plt.annotate('1.645', xy=(1.5,0.14), fontsize=14)
plt.annotate('Z kuat tekan beton',xy=(z_score_csmpa,0.005),xytext=(-50,0.1),
             arrowprops=dict(arrowstyle='->',lw=3, color= "cadetblue"), fontsize=14)
plt.plot([z_score_csmpa],[0], marker="o", color= "cadetblue")
plt.title("Distribusi Kuat Tekan Beton", fontsize=16)
plt.plot(x_axis, y_axis)
plt.legend()
plt.show()
```

Artinya, kita yakin bahwa jika sebuah perencanaan campuran beton dengan semen hanya 200kg, maka kuat tekannya tidak melebihi 60 MPa dengan tingkat kepercayaan 95%.
